```python
import minari

dataset = minari.load_dataset('D4RL/kitchen/mixed-v2')
env = dataset.recover_environment(render_mode='rgb_array')
dataset.set_seed(seed=1)

episode = dataset.sample_episodes(n_episodes=1)[0]
env.reset()
initial_obs = env.render()

for t in range(30):
    obs, reward, _, done, info = env.step(episode.actions[t])

goal_img = env.render()
```

```python
from matplotlib import pyplot as plt

plt.subplot(1, 2, 1)
plt.imshow(initial_obs)
plt.title('Initial State')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(goal_img)
plt.title('Goal State')
plt.axis('off')

plt.tight_layout()
plt.show()
```
